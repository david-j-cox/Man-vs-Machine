# Counts by zip code

```
# Assign a zip code to each observation in batches 
from uszipcode import Zipcode
from uszipcode import SearchEngine
search = SearchEngine(simple_zipcode=True)

ziplist = []

for i in list(range(2473880, 3092351)): # Broke  into chunks of 5 and ran on 5 colab notebooks simultaneously. These are the numbers for the 5th
   zip = search.by_coordinates(nash_data['lat'][i], nash_data['lng'][i], radius=5, returns=1)
   val = [zipcode.zipcode for zipcode in zip]
   ziplist.append(val)
   if (i % 1000 == 0):
     print("Still working...ziplist length=%s" %len(ziplist))
   if (i%10000 == 0):
     print("Another 10,000 down")
     obs_zips = pd.DataFrame(ziplist)
     obs_zips.to_csv('zips_5.csv')
```

```
# Add zipcodes to nash_data df
# Read in data
zip1 = pd.read_csv('zips_1.csv')
zip1 = zip1.drop(['Unnamed: 0'], axis=1)
zip2 = pd.read_csv('zips_2.csv')
zip2 = zip2.drop(['Unnamed: 0'], axis=1)
zip3 = pd.read_csv('zips_3.csv')
zip3 = zip3.drop(['Unnamed: 0'], axis=1)
zip4 = pd.read_csv('zips_4.csv')
zip4 = zip4.drop(['Unnamed: 0'], axis=1)
zip5 = pd.read_csv('zips_5.csv')
zip5 = zip5.drop(['Unnamed: 0'], axis=1)

# Combine into single df
frames = [zip1, zip2, zip3, zip4, zip5]
all_zips = pd.concat(frames)
all_zips.columns = ['zipcodes']
all_zips = all_zips.reset_index()

# Add to nash_data
nash_data['zipcodes'] = all_zips['zipcodes']
```
