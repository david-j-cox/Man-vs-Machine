```python
# Packages!
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display
pd.set_option('display.max_columns', None)
```

```python
# Read in data
data = pd.read_csv('merged_data.csv')
data = data.drop(['Unnamed: 0'], axis=1)
df = data.copy()
df.head(20)
```

## Cox, Sosine, & Dallery (2017) found that aggregating pitches into four categories was most helpful for analysis:
1. Fast & Straight (e.g., four-seam fastball, two-eam fastball)
1. Fast & Breaking (e.g., slider, cutter, sinker)
1. Slow & Straight (e.g., changeup, eephus, knuckleball)
1. Slow & Breaking (e.g., curveball, screwball)

```python
# Break pitches into these four categories. 
fastStr = ['FF', 'FT', ]
fastBr = ['FC', 'FS', 'SI', 'SL']
slowStr = ['CH', 'EP', 'KN']
slowBr = ['CU', 'KC', 'SC']
pitch_category = []
for i in range(len(df)):
    if df.pitch_type[i] in fastStr:
        pitch_category.append('FS')
    if df.pitch_type[i] in fastBr:
        pitch_category.append('FB')
    if df.pitch_type[i] in slowStr:
        pitch_category.append('SS')
    if df.pitch_type[i] in slowBr:
        pitch_category.append('SB')
    else:
        pitch_category.append('UN')

# Add column to df.
df['pitch_category'] = pitch_category
```

```python
# Unique batters
batters = df.batter_id.unique()

# We'll create the pipeline to fit the GME to batter data using an example case
np.count_nonzero(df.batter_id == 572761) # Return total pitches seen by batter 572761
```

```python
# Plot pitch types
pitch_counts = batter_572761['pitch_type'].value_counts()
plt.figure(figsize=(10, 5))
sns.barplot(pitch_counts.index, pitch_counts.values, alpha=0.8, color='red')
plt.ylabel('Number Seen', fontsize=20)
plt.xlabel('Pitch Type', fontsize=20)
```
