```python
# Packages!
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display
pd.set_option('display.max_columns', None)
```

```python
# Read in data
data = pd.read_csv('merged_data.csv')
data = data.drop(['Unnamed: 0'], axis=1)
df = data.copy()
print(len(df))
df.head(10)
```

```python
# Unique batters
batters = df.batter_id.unique()

# We'll create the pipeline to fit the GME to batter data using an example case
np.count_nonzero(df.batter_id == 572761) # Return total pitches seen by batter 572761
```

```python
# Create df for batter 572761
batter_572761 = df[df['batter_id'] == 572761]
batter_572761 = batter_572761.drop(['px','pz','start_speed','end_speed','spin_rate','spin_dir','break_angle',\
                                    'break_length','break_y','ax','ay','az','sz_bot','sz_top','type_confidence',\
                                    'vx0','vy0','vz0','x','x0','y','y0','z0','pfx_x','pfx_z','nasty', 'event_num',\
                                   'o'], axis=1)
batter_572761.head(10)
```

```python
# Plot pitch types
pitch_counts = batter_572761['pitch_type'].value_counts()
plt.figure(figsize=(10, 5))
sns.barplot(pitch_counts.index, pitch_counts.values, alpha=0.8, color='red')
plt.ylabel('Number Seen', fontsize=20)
plt.xlabel('Pitch Type', fontsize=20)
```

```python
# Plot pitch categories
pitch_counts = batter_572761['pitch_category'].value_counts()
plt.figure(figsize=(10, 5))
sns.barplot(pitch_counts.index, pitch_counts.values, alpha=0.8, color='red')
plt.ylabel('Number Seen', fontsize=20)
plt.xlabel('Pitch Type', fontsize=20)
```

```python
# Correlation matrix
corr = batter_572761.corr()
plt.figure(figsize=(10,10))
ax = sns.heatmap(corr, vmin=-1, vmax=1, center=0,cmap=sns.diverging_palette(20, 220, n=200), square=True)
ax.set_xticklabels(ax.get_xticklabels(),rotation=45,horizontalalignment='right')
plt.show()
```

```python
# Need to isolate all instances of (a) swing or not, crossed with (b) pitch category
fs_sw = 0 # fast and straight (fs) and swinging (sw)
fs_nsw = 0 # fast and straight (fs) and no swing (nsw)
fb_sw = 0
fb_nsw = 0
ss_sw = 0
ss_nsw = 0
sb_sw = 0
sb_nsw = 0

swing = ['S', 'F', 'T', 'L', 'W', 'M', 'Q', 'R', 'X', 'D', 'E'] # Labels at https://www.kaggle.com/pschale/mlb-pitch-data-20152018#pitches.csv
no_swing = ['B', '*B', 'C', 'I', 'P', 'H']

# Create counts
for i in list(range(0, 10):
    if (batter_572761.pitch_category[i] == "FS") AND (batter_572761.type[i] in swing):
        fs_sw += 1
    elif (batter_572761.pitch_category[i] == "FS") AND ((batter_572761.type[i] in ) no_swing):
        fs_nsw += 1
    elif (batter_572761.pitch_category[i] == "FB") AND ((batter_572761.type[i] in ) swing):
        fb_sw += 1
    elif (batter_572761.pitch_category[i] == "FB") AND ((batter_572761.type[i] in ) no_swing):
        fb_nsw += 1
    elif (batter_572761.pitch_category[i] == "SS") AND ((batter_572761.type[i] in ) swing):
        ss_sw += 1
    elif (batter_572761.pitch_category[i] == "SS") AND ((batter_572761.type[i] in ) no_swing):
        ss_nsw += 1
    elif (batter_572761.pitch_category[i] == "SB") AND ((batter_572761.type[i] in ) swing):
        sb_sw += 1
    else (batter_572761.pitch_category[i] == "SB") AND ((batter_572761.type[i] in ) no_swing):
        sb_nsw += 1
```
